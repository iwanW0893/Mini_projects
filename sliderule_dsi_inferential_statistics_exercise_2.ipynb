Note that the 'b' and 'w' values in race are assigned randomly to the resumes when presented to the employer.

3. Compute margin of error, confidence interval, and p-value.

```python
diff = np.sum(white_call) - np.sum(black_call)
print(diff)

bs_times = 10000

bs_diff = np.empty(bs_times)

for i in range(bs_times):
    bs_white = np.random.choice(white_call, len(white_call))
    bs_black = np.random.choice(black_call, len(black_call))
    
    bs_diff[i] = np.sum(bs_white) - np.sum(bs_black)

print(np.percentile(bs_diff, [2.5, 97.5]))
```

```python
permute = 100000

concat = np.concatenate((white_call,black_call))

perm_diff = np.empty(permute)
for i in range(permute):
    permutation = np.random.permutation(concat)
    permu_white = permutation[:len(white_call)]
    permu_black = permutation[len(white_call):]
    perm_diff[i] = np.sum(permu_white)-np.sum(permu_black)

print(np.sum(perm_diff>diff)/len(perm_diff))
```
