## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json_df = pd.read_json(r'C:\Users\kenew\Documents\gwaith_cwrs_iwan\data_wrangling_json\data_wrangling_json\data\world_bank_projects.json')
#loaded the json

print(json_df.columns)
#checking the column names

json_df.groupby(['countryname']).size().sort_values(ascending=False).head(10)

#grouping the country names, unpacking and ordering by values to then output the countries with the most projects.
```

```python
#loading and normalizing the dataset of the namecode as described in the json normalizing instructions.
#checking to see what the dataset looks like.

json_string = json.load((open(r'C:\Users\kenew\Documents\gwaith_cwrs_iwan\data_wrangling_json\data_wrangling_json\data\world_bank_projects.json')))
normalized_ver = json_normalize(json_string, 'mjtheme_namecode', ['id'])
print(normalized_ver.head(10))
```

```python
normalized_ver.groupby(['name']).size().sort_values(ascending=False).head(10)

#like before, grouping and unpacking, then sorting the values to see the top 10 most common project names
```

```python
projects = normalized_ver[['code', 'name']]
#selecting the code and name only, to drop the duplicates and get a table of the codes and names.
```

```python
projects = projects.drop_duplicates().dropna(axis=0, how='any')

#dropping the duplicates and na's to obtain a table of the codes and associated names.

print(projects)
```

```python
pd.merge(projects, normalized_ver, left_on='name', right_on='name')

#merging the data together to ensure the names are with their associated codes.
```
